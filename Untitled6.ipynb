# **Data Preprocessing and feature engineering**

# **Import Required Libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
```

# **Load Dataset**

```python
from google.colab import files
uploaded = files.upload()


train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
full_data = pd.concat([train, test], sort=False)
train.head()
test.head()
```

# **Explore the Dataset**

```python
print(train.shape, test.shape)
train.head()
train.describe()
train.info()
```

# **Missing Value Treatment**

```python
# Check missing values
missing = full_data.isnull().sum()
missing[missing > 0].sort_values(ascending=False)

# Drop features with too many missing values (optional and safe)
drop_features = ['Alley', 'PoolQC', 'Fence', 'MiscFeature']
full_data.drop(columns=drop_features, inplace=True, errors='ignore')  # errors='ignore' handles missing columns safely

# Identify column types
categorical_cols = full_data.select_dtypes(include='object').columns
numerical_cols = full_data.select_dtypes(include=['int64', 'float64']).columns

# Fill missing values for categorical columns with mode
for col in categorical_cols:
    if full_data[col].isnull().sum() > 0:
        full_data[col] = full_data[col].fillna(full_data[col].mode()[0])

# Fill missing values for numerical columns with median
for col in numerical_cols:
    if full_data[col].isnull().sum() > 0:
        full_data[col] = full_data[col].fillna(full_data[col].median())

```

# **Feature Engineering**

```python
full_data['TotalSF'] = full_data['TotalBsmtSF'] + full_data['1stFlrSF'] + full_data['2ndFlrSF']

full_data['HouseAge'] = full_data['YrSold'] - full_data['YearBuilt']
full_data['RemodAge'] = full_data['YrSold'] - full_data['YearRemodAdd']

# Simplify quality features
quality_mapping = {
    'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, np.nan: 0
}
full_data['ExterQual'] = full_data['ExterQual'].map(quality_mapping)
full_data['KitchenQual'] = full_data['KitchenQual'].map(quality_mapping)

# Label encode or One-hot encode
label_encodable = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
                   'LotConfig', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType']

for col in label_encodable:
    lbl = LabelEncoder()
    full_data[col] = lbl.fit_transform(full_data[col])

# One-hot encode remaining
full_data = pd.get_dummies(full_data)
full_data

```

# **Feature Scaling**

```python
scaler = StandardScaler()
numerical_cols = full_data.select_dtypes(include=['int64', 'float64']).columns
full_data[numerical_cols] = scaler.fit_transform(full_data[numerical_cols])
```

# **Split to Train and Test**

```python
train_cleaned = full_data.iloc[:len(train)].copy()
test_cleaned = full_data.iloc[len(train):].copy()

# Add target column back to train
train_cleaned.loc[:, 'SalePrice'] = train['SalePrice'].values

X = train_cleaned.drop('SalePrice', axis=1)
y = train_cleaned['SalePrice']

#Train The Given Model

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

#  Ensure test_cleaned has no target column
if 'SalePrice' in test_cleaned.columns:
    test_cleaned = test_cleaned.drop('SalePrice', axis=1)

#  Predict
predictions = model.predict(test_cleaned)

# Create submission file and view output

submission = pd.DataFrame({
    'Id': test['Id'],  # from original test.csv
    'SalePrice': predictions
})

# View output
print(submission.head())
print(submission['SalePrice'].describe())

```

# **Visualize prediction distribution**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(submission['SalePrice'], kde=True)
plt.title('Predicted House Sale Price Distribution')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```
